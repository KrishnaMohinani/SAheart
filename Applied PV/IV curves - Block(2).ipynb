```python
import pandas as pd
import matplotlib.patches as mpatches
from sklearn.linear_model import LinearRegression
import numpy as np
import datetime
%matplotlib inline
import matplotlib.pyplot as plt
import plotly.express as px 
import seaborn as sns
import cufflinks as cf
import plotly as py
py.offline.init_notebook_mode(connected = True)
plt.rcParams.update({'font.size': 32})
cf.go_offline()
sns.set()
```

```python
IVraw = pd.read_csv('IVraw_Block2_2.csv')
IVraw['P'] = IVraw['V']*IVraw['I']
```

```python
IVraw.groupby(['FlashID'])['V','I','P'].max()
```

```python
Unshaded = pd.DataFrame(IVraw[IVraw['FlashID'] == 1611])
Cell_A2_shaded_by_30 = pd.DataFrame(IVraw[IVraw['FlashID'] == 1612])
Cell_A2_shaded_by_60 = pd.DataFrame(IVraw[IVraw['FlashID'] == 1613])
Cell_A2_and_B2_shaded_by_60 = pd.DataFrame(IVraw[IVraw['FlashID'] == 1614])
A2_shaded_by_60_and_C2_shaded_by_30 = pd.DataFrame(IVraw[IVraw['FlashID'] == 1615])
E2_shaded_by_60_and_C2_shaded_by_30 = pd.DataFrame(IVraw[IVraw['FlashID'] == 1616])
Entire_row10_by_30 = pd.DataFrame(IVraw[IVraw['FlashID'] == 1617])
```

```python
fig, ax = plt.subplots( figsize=(20,10))
plt.title('Unshaded',fontsize=20)
plt.plot(Unshaded['V'], Unshaded['I'],linewidth = 4);
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)

plt.xlabel('Voltage (V)',fontsize=20)
plt.ylabel ('Current (A)',fontsize=20)
plt.ylim(0,11)

ax2 = ax.twinx()
ax2.plot(Unshaded['V'], Unshaded ['P'], color = 'red', linestyle = 'dashed', label ='Power curve');
plt.yticks(fontsize=18)
plt.ylabel ('Power (W)',fontsize=20)
fig.legend(loc='upper right',fontsize=18, bbox_to_anchor=(1,1), bbox_transform=ax.transAxes)
plt.ylim(0,250)
```

```python
fig, ax = plt.subplots( figsize=(20,10))
plt.title('Entire row 10 by 30 %',fontsize=20)
plt.plot(Entire_row10_by_30['V'], Entire_row10_by_30['I'], linewidth = 4,  label ='IV curve')
plt.ylim(0,10)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.xlabel('Voltage (V)',fontsize=20)
plt.ylabel ('Current (A)',fontsize=20)

ax2 = ax.twinx()
ax2.plot(Entire_row10_by_30['V'], Entire_row10_by_30 ['P'], color = 'red', linestyle = 'dashed', label ='Power curve');
plt.yticks(fontsize=18)
plt.ylabel ('Power (W)',fontsize=20)
plt.ylim(0,225)
fig.legend(loc='upper right',fontsize=18, bbox_to_anchor=(1,1), bbox_transform=ax.transAxes)
```
