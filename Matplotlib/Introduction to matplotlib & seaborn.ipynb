```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,10),sharex=True)

# 3. Plot data
ax1.plot(x, y)
ax2.plot(x,y)

# 4. Customize plot
ax1.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig('')
```

## Making the most common type of plots using NumPy arrays
Most of figuring out what kind of plot to use is getting a feel for the data, then see what suits it best.

Matplotlib visualizations are built off NumPy arrays. So in this section we'll build some of the most common types of plots using NumPy arrays.

- `line`
- `scatter`
- `bar`
- `hist`
- `subplots()`

To make sure we have access to NumPy, we'll import it as `np`.

```python
evenly_spaced = np.linspace(0,10,100)
evenly_spaced[-5:]
```

```python
# Line plot
fig, ax = plt.subplots()
ax.plot(evenly_spaced,np.square(evenly_spaced));
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(evenly_spaced,np.exp(evenly_spaced));
```

```python
# another Scatter plot
fig, ax = plt.subplots()
ax.scatter(evenly_spaced,np.cos(evenly_spaced));
```

```python
# Bar plot
nut_butter_sales = {'Almond butter' : 10,
                    'Peanut butter' : 8,
                    'Cashew butter' : 12 }
fig, ax = plt.subplots()
ax.bar(nut_butter_sales.keys(),nut_butter_sales.values())
ax.set(title='Nut butter sales',ylabel='Price');

    
```
