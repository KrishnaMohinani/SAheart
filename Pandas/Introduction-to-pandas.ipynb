#### Dealing with missing values

```python
missing_data_car = pd.read_csv('car-sales-missing-data.csv') 
missing_data_car
```

#### Fill the missing values i.e. fillna()

```python
missing_data_car['Odometer'].fillna(0)      #This replaces the NAN values with zero
```

```python
missing_data_car['Odometer'].fillna(missing_data_car['Odometer'].mean())      #This replaces NAN values with the mean value
```

```python
missing_data_car.fillna(0)    #Replaces all NAN values with zero
```

#### We can reassign values with the help of inplace function i.e. inplace = True

```python
missing_data_car['Odometer'].fillna(0,inplace=True)   
```

```python
missing_data_car
```

#### Remove the missing values i.e. dropna()

```python
removed_DF = missing_data_car.dropna()
removed_DF
```

```python
missing_data_car
```

```python
removed_DF
```

#### Building an new column

```python
car_seats = pd.Series([4,5,2,5,4,5,2,5,5,3])
missing_data_car['Seats'] = car_seats
missing_data_car
```

##### We can also create a column using python list but the indexes should match

```python
fuel_economy = range(1,11,1)
missing_data_car['Fuel per 50 KM'] = fuel_economy
missing_data_car
```

#### We can also perform mathematical operations between column and an example is shown below

```python
missing_data_car['Fuel used'] = missing_data_car['Odometer']/100 * missing_data_car['Fuel per 50 KM']
missing_data_car
```

#### Creating a column from a single value

```python
missing_data_car['Number of wheels'] = 4
missing_data_car
```

#### We can also remove a column with the drop function

```python
missing_data_car.drop('Number of wheels',axis=1)         #Axis 1 represents a column
```

```python
missing_data_car['Odometer'] = missing_data_car['Odometer']/1.6
missing_data_car
```

#### Randomize data

```python
shuffled_data = missing_data_car.sample(frac = 1)                 # all data is randomized if frac is 1.
shuffled_data
```

#### Now resetting the randomise data using reset_index

```python
shuffled_data.reset_index(drop = True, inplace = True) 
shuffled_data
```
